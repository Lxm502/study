- 2).`output_dim`ï¼šæ•´æ•°ï¼Œå¯†é›†åµŒå…¥çš„ç»´åº¦ã€‚å³æ¯ä¸ªç´¢å¼•ä¼šè¢«è½¬æ¢æˆå¤šå°‘ç»´çš„å‘é‡ã€‚è¿™æ˜¯ä¸€ä¸ªè¶…å‚æ•°ï¼Œå¸¸è§å€¼åœ¨ 50 åˆ° 300 ä¹‹é—´ï¼Œå–å†³äºä»»åŠ¡å’Œæ•°æ®é›†å¤§å°ã€‚

### å…¶ä»–é‡è¦å‚æ•°ï¼š

- 3).`embeddings_initializer`ï¼šåµŒå…¥çŸ©é˜µçš„åˆå§‹åŒ–å™¨ï¼ˆé»˜è®¤ä¸º '`uniform`' éšæœºå‡åŒ€åˆ†å¸ƒï¼‰ã€‚ä½ å¯ä»¥åœ¨è¿™é‡Œä½¿ç”¨é¢„è®­ç»ƒçš„è¯å‘é‡ï¼ˆå¦‚ `GloVe`ã€`Word2Vec`ï¼‰æ¥åˆå§‹åŒ–ï¼Œè¿™é€šå¸¸èƒ½æ˜¾è‘—æå‡æ¨¡å‹æ•ˆæœã€‚

- 4).`mask_zero`ï¼šå¸ƒå°”å€¼ã€‚æ˜¯å¦å°†ç´¢å¼• 0 è§†ä¸ºéœ€è¦è¢«â€œæ©ç›–â€çš„ç‰¹æ®Šå€¼ï¼ˆé€šå¸¸æ˜¯å¡«å……ç¬¦ `<pad>`ï¼‰ã€‚`StringLookup` å±‚é€šå¸¸å°† `mask_token` æ˜ å°„ä¸ºç´¢å¼• 0ã€‚å°†å…¶è®¾ç½®ä¸º `True` åï¼Œæ¨¡å‹åç»­çš„å±‚ï¼ˆå¦‚ `RNN`, `Transformer`ï¼‰ä¼šè‡ªåŠ¨è·³è¿‡è¿™äº›ä½ç½®ï¼Œä¸å‚ä¸è®¡ç®—ã€‚éå¸¸é‡è¦ï¼

- 5).`input_length`ï¼šæ•´æ•°ã€‚è¾“å…¥åºåˆ—çš„å›ºå®šé•¿åº¦ã€‚å½“ä½ åé¢æ¥ `Flatten` å’Œ `Dense` å±‚æ—¶ï¼Œéœ€è¦æŒ‡å®šè¿™ä¸ªå‚æ•°ä»¥ä¾¿å±‚èƒ½è®¡ç®—å‡ºè¾“å‡ºçš„å›ºå®šå¤§å°ã€‚å¦‚æœåé¢æ˜¯ RNN æˆ–å…¨å±€æ± åŒ–å±‚ï¼Œåˆ™é€šå¸¸ä¸éœ€è¦ã€‚

## å››ã€å®Œæ•´å·¥ä½œæµç¨‹ç¤ºä¾‹
è¿™æ˜¯ä¸€ä¸ªæ ‡å‡†çš„æ–‡æœ¬åˆ†ç±»æµç¨‹ï¼Œå±•ç¤ºäº† StringLookup å’Œ Embedding å¦‚ä½•ååŒå·¥ä½œã€‚

```python
import tensorflow as tf

# 1. ç¤ºä¾‹æ–‡æœ¬æ•°æ®
texts = [
    "I love my dog",
    "I love my cat",
    "Do you think my dog is amazing?",
    "My cat is very cute"
]
labels = [1, 1, 1, 0] # 1 for positive, 0 for negative

# 2. æ–‡æœ¬æ ‡å‡†åŒ–å’Œåˆ†è¯
# å°†ä¸€ä¸ªå­—ç¬¦ä¸²æˆ–å­—ç¬¦ä¸²æ‰¹æ¬¡ï¼ˆBatchï¼‰æ˜ å°„åˆ°æ•´æ•° token çš„åºåˆ—æˆ–å¯†é›†ï¼ˆDenseï¼‰æ•°å€¼è¡¨ç¤ºï¼ˆå¦‚ TF-IDFï¼‰ï¼Œä»è€Œä½¿å…¶å¯ä»¥è¢«ç¥ç»ç½‘ç»œå¤„ç†ã€‚
vectorize_layer = tf.keras.layers.TextVectorization(
    max_tokens=10,       # åªä¿ç•™å‰10ä¸ªæœ€é¢‘ç¹çš„è¯
    output_mode='int',   # è¾“å‡ºæ•´æ•°ç´¢å¼•
    output_sequence_length=5 # ç»Ÿä¸€åºåˆ—é•¿åº¦ä¸º5ï¼ˆä¸è¶³å¡«å……ï¼Œè¿‡é•¿æˆªæ–­ï¼‰
)
vectorize_layer.adapt(texts) # å­¦ä¹ è¯æ±‡è¡¨
vocab = vectorize_layer.get_vocabulary()
print("Vocabulary:", vocab)
# è¾“å‡ºå¯èƒ½: ['', '[UNK]', 'my', 'love', 'dog', 'cat', 'i', 'you', ...]
# ç´¢å¼•: 0->'', 1->'[UNK]', 2->'my', 3->'love', 4->'dog', 5->'cat'

# 3. æ„å»ºæ¨¡å‹
model = tf.keras.Sequential([
    tf.keras.Input(shape=(1,), dtype=tf.string), # è¾“å…¥åŸå§‹å­—ç¬¦ä¸²
    vectorize_layer,                             # æ–‡æœ¬ -> æ•´æ•°ç´¢å¼•
    tf.keras.layers.Embedding(
        input_dim=len(vocab),    # è¯æ±‡è¡¨å¤§å°
        output_dim=64,           # åµŒå…¥ç»´åº¦ï¼Œè¶…å‚æ•°
        mask_zero=True           # å¯ç”¨æ©ç ï¼Œå¿½ç•¥ç´¢å¼•0ï¼ˆå¡«å……ç¬¦ï¼‰
    ), # è¾“å‡ºå½¢çŠ¶: (None, 5, 64)
    # æ¥ä¸‹æ¥å¯ä»¥æ¥å„ç§å±‚ï¼Œä¾‹å¦‚ï¼š
    tf.keras.layers.LSTM(128),   # LSTMå±‚ä¼šè‡ªåŠ¨å¤„ç†æ©ç 
    # tf.keras.layers.GlobalAveragePooling1D(), # æˆ–è€…ç®€å•æ± åŒ–
    tf.keras.layers.Dense(1, activation='sigmoid') # äºŒåˆ†ç±»è¾“å‡ºå±‚
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.summary()

# 4. è®­ç»ƒæ¨¡å‹
# æ³¨æ„ï¼šè¾“å…¥æ˜¯åŸå§‹å­—ç¬¦ä¸²æ•°ç»„ï¼Œæ¨¡å‹å†…éƒ¨ä¼šè‡ªåŠ¨å®Œæˆç´¢å¼•åŒ–å’ŒåµŒå…¥
history = model.fit(texts, labels, epochs=5)
```

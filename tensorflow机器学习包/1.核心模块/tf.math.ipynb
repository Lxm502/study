## ğŸ¯ ä¸‰ã€å®æˆ˜ç¤ºä¾‹ï¼šSoftmax + Log + Argmax

```python
logits = tf.constant([2.0, 1.0, 0.1])
probs = tf.math.softmax(logits)
log_probs = tf.math.log(probs)
pred = tf.math.argmax(probs)

print("Probabilities:", probs.numpy())
print("Log-Probabilities:", log_probs.numpy())
print("Predicted class:", pred.numpy())
```

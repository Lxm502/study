# tf.math 
是 TensorFlow 中用于执行数学运算的模块，是张量计算的基础工具之一，功能类似于 NumPy 中的 `np.math`，但用于 `TensorFlow` 的张量（tf.Tensor）对象，且支持自动微分、GPU 加速等特性。

## 🔹 一、导入与基本使用
import tensorflow as tf

# tf.math 不需要单独导入，直接用 tf.math.xxx 或 tf.xxx 即可

---
## ✅ 二、常用函数分类汇总
下面按功能分类列出 tf.math 的常用函数：
### 🔢 1. 基本数学运算
| 函数                       | 说明 | 示例                            |
| ------------------------ | -- | ----------------------------- |
| `tf.math.add(x, y)`      | 加法 | `tf.math.add(1, 2)` → 3       |
| `tf.math.subtract(x, y)` | 减法 | `tf.math.subtract(5, 3)` → 2  |
| `tf.math.multiply(x, y)` | 乘法 | `tf.math.multiply(2, 3)` → 6  |
| `tf.math.divide(x, y)`   | 除法 | `tf.math.divide(5, 2)` → 2.5  |
| `tf.math.floordiv(x, y)` | 整除 | `tf.math.floordiv(5, 2)` → 2  |
| `tf.math.mod(x, y)`      | 取余 | `tf.math.mod(5, 2)` → 1       |
| `tf.math.pow(x, y)`      | 幂  | `tf.math.pow(2.0, 3.0)` → 8.0 |

---
### 📏 2. 统计类函数
| 函数                             | 说明     |
| ------------------------------ | ------ |
| `tf.math.reduce_sum(x, axis)`  | 求和     |
| `tf.math.reduce_mean(x, axis)` | 求平均    |
| `tf.math.reduce_max(x, axis)`  | 求最大值   |
| `tf.math.reduce_min(x, axis)`  | 求最小值   |
| `tf.math.argmax(x, axis)`      | 最大值的索引 |
| `tf.math.argmin(x, axis)`      | 最小值的索引 |

- x = tf.constant([[1, 2], [3, 4]])
- tf.math.reduce_sum(x)       # 10
- tf.math.reduce_mean(x, axis=0)  # [2, 3]

---
### 📐 3. 初等函数（常见数学函数）
| 函数                               | 说明           |
| -------------------------------- | ------------ |
| `tf.math.abs(x)`                 | 绝对值          |
| `tf.math.square(x)`              | 平方           |
| `tf.math.sqrt(x)`                | 平方根          |
| `tf.math.log(x)`                 | 自然对数 ln(x)   |
| `tf.math.exp(x)`                 | e 的 x 次幂     |
| `tf.math.sin(x)` / `cos` / `tan` | 三角函数         |
| `tf.math.tanh(x)`                | 双曲正切函数       |
| `tf.math.sigmoid(x)`             | sigmoid 激活函数 |
| `tf.math.softmax(x)`             | softmax 激活函数 |

- x = tf.constant([1.0, 2.0, 3.0])\\
- tf.math.softmax(x)  # 输出 softmax 结果

---
### 🔁 4. 比较运算函数（返回布尔张量）
| 函数                            | 说明   |
| ----------------------------- | ---- |
| `tf.math.equal(x, y)`         | 是否相等 |
| `tf.math.not_equal(x, y)`     | 是否不等 |
| `tf.math.less(x, y)`          | 小于   |
| `tf.math.less_equal(x, y)`    | 小于等于 |
| `tf.math.greater(x, y)`       | 大于   |
| `tf.math.greater_equal(x, y)` | 大于等于 |

- a = tf.constant([1, 2, 3])
- b = tf.constant([2, 2, 2])
- tf.math.greater(a, b)  # [False, False, True]

---
### 🔀 5. 逻辑运算（布尔型）
| 函数                          | 说明 |
| --------------------------- | -- |
| `tf.math.logical_and(x, y)` | 与  |
| `tf.math.logical_or(x, y)`  | 或  |
| `tf.math.logical_not(x)`    | 非  |

---
## 🎯 三、实战示例：Softmax + Log + Argmax

```python
logits = tf.constant([2.0, 1.0, 0.1])
probs = tf.math.softmax(logits)
log_probs = tf.math.log(probs)
pred = tf.math.argmax(probs)

print("Probabilities:", probs.numpy())
print("Log-Probabilities:", log_probs.numpy())
print("Predicted class:", pred.numpy())
```

## 📌 四、TensorFlow 与 NumPy 的互通性
| NumPy     | TensorFlow     |
| --------- | -------------- |
| `np.add`  | `tf.math.add`  |
| `np.log`  | `tf.math.log`  |
| `np.sqrt` | `tf.math.sqrt` |

但 TensorFlow 的版本更适用于：\
--  自动求导\
--  GPU/TPU 加速\
--  构建计算图（tf.function 中可用）
